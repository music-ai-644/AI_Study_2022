```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# 절댓값 함수 
x = np.linspace(-10, 10, 100)
y = np.abs(x)
plt.plot(x, y)
plt.grid()
```

```python
# 제곱 함수 
x = np.linspace(-3, 3, 100)
y = x ** 2
plt.plot(x, y)
plt.grid()
```

```python
# 절댓값, 제곱 함수 
x = np.linspace(-3, 3, 100)
y = x ** 2
y2 = np.abs(x)
plt.plot(x, y)
plt.plot(x, y2, 'r')
plt.grid()
```
